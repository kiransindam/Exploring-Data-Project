# (8) Handling outliers

```python
factor=4
up=df['Record ID'].mean()+df['Record ID'].std()*factor
dw=df['Record ID'].mean()-df['Record ID'].std()*factor
new1=df[(df['Record ID']<up)&(df['Record ID']>dw)]
```

```python
new1.info()
```

```python
up1=df['Record ID'].quantile(.98)
dw1=df['Record ID'].quantile(.05)
new=df[(df['Record ID']<up1)&(df['Record ID']>dw1)]
```

# (9) Binning

```python
s=df['Record ID'].quantile(.25)
s1=df['Record ID'].quantile(.50)
s2=df['Record ID'].quantile(.75)
s3=df['Record ID'].quantile(1)
df['Record ID_bin']=pd.cut(df['Record ID'],bins=[0,s,s1,s2,s3],labels=['low','medium','hight','veryhight'])
```

```python
df[['Record ID','Record ID_bin']]
```

# (11) Data Split

```python
new = df["Aircraft: Make/Model"].str.split("-", n = 1, expand = True)
df['Aircraft:Model'] = new[0]
new.head()
```

```python
df['Aircraft:Model'] = new[1]
df.head()
```

```python

list1=['B','C','DC','A','NW','HA','T','UA']
list2=['BUSINESS','CONTINENTAL','DELTA ','AMERICAN ','NORTHWEST','HAWAIIAN ','TRANS WORLD','UNITED AIRLINES']
df['Aircraft:Model'].replace(list1,list2,inplace=True)
df.loc[~df['Aircraft:Model'].isin(list2),'Aircraft:Model']='others'
g=df.groupby('Aircraft:Model').size()
print(g)
```
